```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = 10
learning_rate = 0.2
max_iter = 100
```

```python
loss_func = lambda x: (x-3)**2 +10
##로스 펑션
```

```python
gradient = lambda x:2*x-6
# 그래디언트 펑션: 손실함수의 1차 미분값
```

```python
list1 = []
list2 = []
```

```python
for i in range(max_iter):
    x = x - learning_rate * gradient(x)
    list1.append(x)
    list2.append(loss_func(x))
    print('손실함수값(', x,')', loss_func(x))
```

```python
x1 = np.linspace(0.0,10.0)
y1 = loss_func(x1)
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(list1, list2, '*')
```

```python
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(x, y)
Z = X**2 + y**2
```
